```python
import mysql.connector
import pandas as pd
db_connection = mysql.connector.connect(user='root', password='Gregory2',host='localhost',
                                     port='3306', database='pbl3210 db')
```

```python
stockidtab=pd.read_sql_query("SELECT * FROM STOCK_ID",db_connection)
stockpricetab=pd.read_sql_query("SELECT * FROM STOCKPRICE",db_connection)
```

```python
print(stockidtab)
print(stockpricetab)
for i in stockidtab:
    print(i)
```

```python
word=input("Please input the Ticker of the stock you are looking for \n")
id_value=0
for i in stockidtab.index:
    if stockidtab.at[i,"Ticker Symbol"] ==word:
        id_value=stockidtab.at[i,"Stock ID"]
```

```python
list100=[]
for i in stockpricetab.index:
    if stockpricetab.at[i,"stock_id"]==id_value:
        list100.append((stockpricetab.at[i,"Date"],stockpricetab.at[i,"Price"]))
    elif stockpricetab.at[i,"stock_id"]>id_value:
        break
```

```python
print(list100)
```

```python
sp=pd.DataFrame(list100,columns=["Date", "Price"])
```
